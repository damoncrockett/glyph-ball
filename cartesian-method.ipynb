```python
popfactor = 0.0001
```

```python
# we'll always assume a centering at (0,0)
# x2 + y2 = r2
```

```python
from shapely.geometry import Point
```

```python
def genpts(r):
    
    pts = []
    roster = np.arange(-r,r,popfactor)
    
    for i in range(len(roster)):
        
        a = np.random.choice(roster,1)[0]
        b = np.sqrt( r*r - a*a )
        
        if np.random.choice(["x","y"],1)[0]=="x":
            if np.random.choice([1,-1],1)[0]==1:
                pt = Point(a,b)
            else:
                pt = Point(a,-b)
        else:
            if np.random.choice([1,-1],1)[0]==1:
                pt = Point(b,a)
            else:
                pt = Point(-b,a)
    
        pts.append(pt)
        
    return pts
```

```python
pts = genpts(1)
```

```python
x = [item.x for item in pts]
y = [item.y for item in pts]
```
